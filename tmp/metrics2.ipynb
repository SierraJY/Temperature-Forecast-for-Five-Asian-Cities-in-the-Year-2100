```python
tf_df = pd.read_csv(f"/Users/juyeonm1/Desktop/CAPSTONE/TF_model/TF_forecast/for_mape_seoul.csv")[['DateTime','Monthly Anamoly']]
tf_df["ds"] = pd.to_datetime(tf_df["DateTime"]).dt.to_period('M')
tf_df = tf_df.loc[tf_df.loc[:,'DateTime'] >='2015-01-01']
tf_df = tf_df.loc[tf_df.loc[:,'DateTime'] <='2020-12-01']
tf_df
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

MAPE = mean_absolute_percentage_error(all_df["Monthly Anamoly"], tf_df["Monthly Anamoly"])
RMSE = mean_squared_error(all_df["Monthly Anamoly"], tf_df["Monthly Anamoly"]) ** 0.5

tf_df = pd.DataFrame({"Model" : ["tf"], "MAPE%" : [MAPE*100], "RMSE" : [RMSE]})
tf_df
```
