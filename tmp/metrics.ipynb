```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import dates
```

```python
metric_df = pd.DataFrame(columns=["Model", "MAPE%", "RMSE"])
metric_df
```

```python
all_df = pd.read_csv("./data/all_cities_month_temp_df.csv")
all_df = all_df.loc[all_df["city_num"] == 0]
all_df = all_df.loc[all_df['DateTime'] >= '2015-01-01']
display(all_df)
```

```python
nbeats_df = pd.DataFrame({"Model" : ["NBEATS"], "MAPE%" : [8.46], "RMSE" : [0.037]})
nbeats_df
```

```python
tf_df = pd.DataFrame({"Model" : ["TRANSFORMER"], "MAPE%" : [45.7], "RMSE" : [1.6]})
tf_df
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
prophet_df = pd.read_csv(f"./PROPHET_model/prophet_forecast/prophet_forecast_seoul.csv")[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
prophet_df["ds"] = pd.to_datetime(prophet_df["ds"]).dt.to_period('M')
prophet_df = prophet_df.loc[prophet_df.loc[:,'ds'] >='2015-01-01']
prophet_df = prophet_df.loc[prophet_df.loc[:,'ds'] <='2020-12-01']
prophet_df
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

MAPE = mean_absolute_percentage_error(all_df["Monthly Anamoly"], prophet_df["yhat"])
RMSE = mean_squared_error(all_df["Monthly Anamoly"], prophet_df["yhat"]) ** 0.5

ph_df = pd.DataFrame({"Model" : ["PROPHET"], "MAPE%" : [MAPE*100], "RMSE" : [RMSE]})
ph_df
```

```python
metric_df = pd.concat([metric_df, nbeats_df, tf_df, ph_df], axis=0)
metric_df
```
